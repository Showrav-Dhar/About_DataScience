```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits # inbuilt dataset of sklearn
```

```python
digits = load_digits()## stored all the data in the digits variable
```

## X - > digits[data]
## Y ->  digits[target]

### We will use -> data and target to train our model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=10000)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

### Confusion Matrix - 

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,y_predicted)
cm
```

## Visualizing confusion_matrix using seaborn

```python
import seaborn as sn
plt.figure(figsize=(5,5))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
