```python
import pandas as pd 
df = pd.read_csv('homeprices_for_oneHotEncoding.csv')
df
```

### Creating dummy variable columns

```python
dummy_df = pd.get_dummies(df.town)
dummy_df
```

```python
new_merged_df = pd.concat([df,dummy_df],axis='columns')
new_merged_df
```

### Now you need to drop the original Town coloumn,Because we already have dummy varible column describing the towns,after town coloumn , 

#### you also have to drop any one of the dummy variable column.its like if you have n dummy variable column you have to delete 1 from them and you work with n-1 dummy coloumns


##### so we are dropping main town column and west windsor column

```python
final_df = new_merged_df.drop(['town','west windsor'],axis='columns')
final_df
```

#### although sklearn linear regression works even if you don't drop dummy variable column

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# fix X and Y for the Model

# here X is area,and town names and Y is price

X = final_df.drop(['price'],axis='columns') # to get only the area and town names we made a df without price
X
```

```python
Y = final_df.price # as we only need price, so just took the price from the df
Y
```

```python
# now train the model

model.fit(X,Y) # its training the model
```

```python
## now prediction 

# predict price of 2800 sqft area in robinsville

# pass the data as you trained -> 
# X = area, monroe township, robinsville
model.predict([[2800,0,1]])



# here area = 2800, first 0 means not monroe township, second 1 means robinsville

# so we are predicting the price of 2800 sqft area in robinsville

```

```python
## now prediction 

# predict price of 3400 sqft area in west winsdor

# pass the data as you trained -> 
# X = area, monroe township, robinsville
model.predict([[3400,0,0]])



# here area = 3400, first 0 means not monroe township, second 0 means not robinsville

# so we are predicting the price of 3400 sqft area inwest winsdor

```

```python
# to check the accuracy of model

model.score(X,Y)
```

```python
# it means the model is 95% accurate
```
