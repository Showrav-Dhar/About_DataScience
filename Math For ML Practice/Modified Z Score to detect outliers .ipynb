```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('movie_revenues.csv')
df.head()
```

```python
df.revenue.describe()
```

```python
# to show revenues in million we are adding a new column to the dataframe
df['revenue_mil'] = df['revenue'].apply(lambda x : x/100000)
df.revenue_mil.describe()
# now they are in millions
```

```python
df.head(2)
# check the last column 
# 27879.65 million revenue for the movie 'avatar'
```

```python
# now calculating z score 
df.revenue_mil.describe()
```

```python
mean = df.revenue_mil.mean()
mean
```

```python
std = df.revenue_mil.std()
std
```

```python
# now adding new column that will contain z-score of each row
```

```python
def get_z_score(value,mean,std):
    return (value - mean)/std
```

```python
df['z_score'] = df.revenue_mil.apply(lambda x : get_z_score(x,mean,std))
df.head(2)#check the last coulmn

# lambda x: This defines the parameter of the anonymous function.It represents each value in the revenue_mil column.

# get_z_score(x, mean, std): This is the function call within the lambda function. 
# It takes the current value (x), the mean, and the standard deviation as arguments and returns the z-score.

# apply(lambda_function): This applies the lambda function to each element in the revenue_mil column, creating a new z_score column.
```

```python
# to remove outliers, anything with z-score > 3 is outlier 
```

```python
df[df.z_score>3]
df.head(3)

# here we can see that avatar's revenue is showing outlier.but it is possible and it is real data that, 
# the movie made that much profit.so this is not a outlier in term of real world data
```

```python
# now let's try with modified z-score
```

```python
# median absolute deviation - mad
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median) # absolute
    MAD = np.median(diff)
    
    return MAD
```

```python
MAD = get_mad(df.revenue_mil)
median = np.median(df.revenue_mil)
MAD,median
```

```python
def get_modified_z_score(x,median,MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(27879,median,MAD)
# avatar revenue -> 27879.65087
```

```python
df['mod_z_score'] = df.revenue_mil.apply(lambda x : get_modified_z_score(x,median,MAD))
df.head(2)
```

```python
# to detect outlier - anything greater than modified z-score of 3.5 is outlier
df[df.mod_z_score>3.5]
df.head(3)
```
